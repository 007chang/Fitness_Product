Salary starts from 30000 and goes beyond 100000. there are so many outliers as well, lets try to convert in categorical data, in to low income, middle income and High Income to understand the more about our existing customer

```python
def inbr(sl_br):
    
    if sl_br<44058:
        return "Low Income"
    elif sl_br<58668:
        return "Middle Income"
    else:
        return "High Income"
```

```python
aerofit['Income_Bracket'] = aerofit['Income'].apply(inbr)
```

```python
aerofit['Income_Bracket'].value_counts()
```

50 % people are in Middle Income category and 25 in high and 25 in low income

```python
lower_income = aerofit["Income"].quantile(0.05)
upper_income = aerofit["Income"].quantile(0.95)
clipped_income = aerofit['Income'].clip(lower_income,upper_income)
sns.boxplot(clipped_income,color='Red')
```

after clipping minimum 5% and maximum 5% our Income bracket will look like as above
